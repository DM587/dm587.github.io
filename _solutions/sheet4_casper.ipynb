```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx


# Functions
def permute(M, P):
    return P@(P@M).transpose()

def make_perm_matrix(l):
    P = np.zeros((len(l),len(l)))
    for i,j in enumerate(l):
        P[i,j] = 1
        P[j,i] = 1
    return P

def drawmatrix(M):
    plt.figure(np.random.randint(20000))
    nx.draw(nx.Graph(M))
```
